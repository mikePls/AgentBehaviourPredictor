```python
import pickle
import matplotlib.pyplot as plt
```

```python
total_scores = {i:0  for i in range(32) if i%2==0} # Accumulated scores
dist_scores = {i:0  for i in range(32) if i%2==0} # Distance scores
grip_scores = {i:0  for i in range(32) if i%2==0} # Grip setting
vis_scores = {i:0  for i in range(32) if i%2==0}
```

```python

for i in range(1, 101):
    with open(f'saved_data/probs_data{i}.pkl', 'rb') as f:
        probs_data = pickle.load(f)

        target = probs_data['target']
        keys = [key for key in probs_data.keys() if key !='target']

        for key in keys:
            max_prob = max(probs_data[key]['Total probabilities'], key=probs_data[key]['Total probabilities'].get)
            if max_prob == target:
                total_scores[key] += 1
            max_prob = max(probs_data[key]['Probabilities according to distance'], key=probs_data[key]['Probabilities according to distance'].get)
            if max_prob == target:
                dist_scores[key] += 1
            max_prob = max(probs_data[key]['Probabilities according to grip setting'], key=probs_data[key]['Probabilities according to grip setting'].get)
            if max_prob == target:
                grip_scores[key] += 1
            max_prob = max(probs_data[key]['Probabilities according to number of visibles'], key=probs_data[key]['Probabilities according to number of visibles'].get)
            if max_prob == target:
                vis_scores[key] += 1
            
```
